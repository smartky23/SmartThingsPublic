```python
df = pd.read_excel('/content/data_1117_20250917.xlsx')
```

```python
print(df.head(5))
```

```python
# search_stock_info 함수가 df 데이터를 사용하도록 수정
def search_stock_info():
    query = input("종목명 또는 종목코드를 입력하세요: ")

    # 종목코드로 검색
    result = df[df['종목코드'] == query]
    if not result.empty:
        print(f"종목코드 {query}의 종목명은: {result['종목명'].iloc[0]} 입니다.")
        return

    # 종목명으로 검색
    result = df[df['종목명'].str.contains(query, na=False)]
    if not result.empty:
        print(f"종목명 '{query}'에 해당하는 종목코드는: {result['종목코드'].iloc[0]} 입니다.")
    else:
        print("해당하는 정보를 찾을 수 없습니다.")

# 예시 데이터프레임 df는 이제 사용되며, search_stock_info 함수는 df를 사용합니다.
```

```python
search_stock_info()
```
